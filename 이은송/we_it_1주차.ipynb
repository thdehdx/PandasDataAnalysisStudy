## Series : 1차원 배열

```python
import pandas as pd

# Series를 생성할 때, index를 전달
tup_data = ('영인','2010-05-01','여',True)
sr= pd.Series(tup_data, index = ['이름','생년월일','성별','학생여부'])

print(sr)

# Series의 원소를 인덱스로 출력
print(sr[['이름','생년월일']])


# ★정수로 원소 호출 시 끝 번호 포함되지 않음
print(sr[0:2])          # 성별(인덱스 : 2)은 출력되지 않음
```

## 3. 시리즈, 데이터프레임 연산

    - 사칙연산에 대해 각 원소에 대해 수행
    - 데이터프레임,시리즈 간의 연산은 일치하는 인덱스에 대해서 수행하고 일치하지 않으면 유효하지 않으므로 NaN으로 출력

### Series 연산

```python
import numpy as np
import pandas as pd

student1 = pd.Series({'국어':np.nan, '영어':80, '수학':90})
student2 = pd.Series({'수학':80,'국어':90})

addition = student1 + student2

result = pd.DataFrame(addition)
result
```

#### NaN 대신 0 을 채우고 싶다면,
    Dataframe.add(Datafrme, fill_value=0)

```python
import numpy as np
import pandas as pd

student1 = pd.Series({'국어':np.nan, '영어':80, '수학':90})
student2 = pd.Series({'수학':80,'국어':90})

addition = student1.add(student2,fill_value = 0)

result = pd.DataFrame(addition)
result
```

### 4. 웹 스크래핑

```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
```

```python
url = "https://en.wikipedia.org/wiki/List_of_American_exchange-traded_funds"
resp = requests.get(url)
```

```python
soup = BeautifulSoup(resp.text,'lxml')
rows = soup.select('div>ul>li')
```

```python
etfs = {}
for row in rows:
    
    try : 
        etf_name = re.findall('^(.*)\(NYSE',row.text)
        etf_market = re.findall('\((.*)\|', row.text)
        etf_ticker = re.findall('NYSE Arca\|(.*)\)',row.text)
        
        if (len(etf_ticker)>0) & (len(etf_market)>0):
            etfs[etf_ticker[0]] = [etf_market[0], etf_name[0]]
            
    except AttributeError as err:
        pass
```

```python
print(etfs)
print('\n')

df = pd.DataFrame(etfs)
print(df)
```
